### Анализ данных

Выведем первую десятку городов по количеству рейсов.

```python
cities.nlargest(10, columns='average_flights')
```

Первые семь строчек рейтинга -- это города-милионники, а в Сочи в сентябре бархатный сезон. Любопытно, что не самые крупные Ульяновск и Брянск обогнали остальные миллионники. Возможно, дело в том, что они связаны авиасообщением с популярными курортами. Как бы там ни было, если компания "F9" еще не представлена на каких-то из этих направлений, в ее интересах это исправить. 

Построим графики. Сначала количество полетов в зависимости от модели самолета.

```python
models.sort_values(by='flights_amount', ascending=False, inplace=True)
models.plot(x='model', y='flights_amount', kind='bar')
```

На первом месте оказался самолет, который обычно используется в региональной и бизнес-авиации, на втором -- ближнемагистральный лайнер на 50 посадочных мест, на третьем -- постоянно мелькающий в новостях "Суперджет". Остальные пять моделей принадлежат мировым лидерам в гражданской авиации Boeing и Airbus. По числу полетов они сильно отстают от тройки лидеров: одни только "Цессны" совершили в сентябре 2018 г. больше полетов, чем "Боинги" и "Аэробусы" всех моделей вместе взятые. 

Для неспециалиста самое удвительное в этом графике -- первенство ближнемагистральных самолетов малой вместимости. В интересах компании F9 более подробно изучить направления, на которых используются такие самолеты, и подумать о том, чтобы какие-то из них заменить на более крупные, если позволяет инфраструктура. Не исключено, что это позволит снизить цену билетов и увеличить выручку компании за счет роста продаж.

Теперь построим график среднего количества рейсов, прибывавших в города за один день в сентябре 2018 года.

```python
cities_sorted = cities.sort_values(by='average_flights', ascending=False)
cities_sorted.plot(x='city', y='average_flights', figsize=(20,10), kind='bar')
```

На этом графике видно главным образом исключительное положение Москвы среди всех остальных попавших в выборку городов. Что покажет такой же график без Москвы?

```python
cities_sorted.query('city != "Москва"').plot(x='city', y='average_flights', figsize=(20,10), kind='bar', grid=True)
```

В выборке около 30 городов, куда в сентябре 2018 года прилетало в среднем 5 рейсов или больше. Если исключить первую десятку, получим список направлений, на которые "F9" стоит обратить внимание во вторую очередь.

Наконец, выведем тот же график по первым десяти городам.

```python
cities.nlargest(10, columns='average_flights').plot(x='city', y='average_flights', kind='bar')
```

Даже здесь Москва лидирует с очень большим отрывом. Скорее всего, дело не только в международных рейсах: известно, что через Москву между собой связаны некоторые очень удаленные (от Москвы, но далеко не всегда друг от друга) регионы. С одной стороны, московское направление самое прибыльное, и компании "F9", если она летает в Москву, стоит за него держаться. С другой стороны, на всех остальных направлениях должна быть значительно ниже конкуренция.

Теперь с помощью библиотеки `scipy` проверим гипотезу о взаимосвязи между фестивалями и спросом на авибилеты.

Нулевую гипотезу формулируем так, чтобы она означала сохранение статуса-кво: "Средний спрос на билеты во время фестивалей **не отличается** от среднего спроса на билеты в обычное время". 

Тогда альтернативная гипотеза будет такой: "Средний спрос на билеты во время фестивалей **отличается** от среднего спроса на билеты в обычное время". 

Пороговое значение примем равным 0,05. Поскольку сравниваемые выборки представляют собой части одной и той же генеральной совокупности, параметр `equal_var` трогать не будем.

В таблице `festivals` всего десять строк, поэтому мы столкнемся с проблемой очень маленьких выборок. Чтобы быть уверенными, что данных хватит для обоснованных выводов, нам нужно убедиться, что выборки распределены нормально. Для этого сначала применим критерий Шапиро-Уилка. Пороговое значение снова примем равным 0,05.

```python
sample_1 = festivals.query('festival_name.isna()').ticket_amount
sample_2 = festivals.query('festival_name.notna()').ticket_amount

stats.shapiro(sample_1)[1]
```

```python
stats.shapiro(sample_2)[1]
```

В обоих случаях p-value больше 0,05 (хотя во второй выборке, где всего три значения, и несильно), оснований отбрасывать эту вспомогательную нулевую гипотезу (о нормальности распредления выборок) нет. Значит, можно применять критерий Стьюдента.

```python
results = stats.ttest_ind(sample_1, sample_2)
print('p-значение: ', results.pvalue)
```

При заданном уровне статистической достоверности нулевая гипотеза остается в силе. Спрос на авиабилеты не коррелирует с фестивалями. 

Из этого могут следовать два вывода: либо эти фестивали не настолько популярны, чтобы ради них люди готовы были отправиться в дальнюю поездку (авиация в России используется в основном для перемещения на дальние расстояния). Либо, что менее вероятно, аудитория этих фестивалей предпочитает добираться до места другими видами транспорта. Во втором случае компания "F9" могла бы побороться с железнодорожными и автобусными перевозчиками, но здравый смысл подсказывает, что дело в первой причине: фестивали "Park Live" и "Пикник Афиши" посещают в основном жители московского региона. Что до слета видеоблогеров, который в 2018 году прошел в Санкт-Петербурге, то тем более не ясно, зачем туда ехать, если можно просто посмотреть в видеоблоге.

Для авиакомпании "F9" это означает, что дополнительные рейсы в дни фестивалей, скорее всего, не окупятся.
