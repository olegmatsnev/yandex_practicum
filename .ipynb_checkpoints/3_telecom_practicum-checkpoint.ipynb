На практике операторы обычно продают дополнительный трафик не помегабайтно, а порциями, но в условии об этом ничего не сказано, так что будем считать, что действует помегабайтная тарификация.

Посмотрим, как данные, которые нам выдали для анализа, распределены по двум исследуемым тарифам.

```python
calls.name = 'calls'
messages.name = 'messages'
internet.name = 'internet'

def count_entries(df_list): # функция для подсчета распределения строк между двумя тарифами
    tariff_stats = users.groupby('tariff').count()[['user_id']]
    for df in df_list:
        df_temp = df.set_index('user_id').join(users.set_index('user_id').tariff.to_frame())
        df_grouped = df_temp.groupby('tariff').count().iloc[:, 0]
        tariff_stats = tariff_stats.join(df_grouped.to_frame(), rsuffix='_' + df.name)
    tariff_stats.rename(columns={'user_id': 'users', 'id': 'messages', 
                                 'id_calls': 'calls', 'id_internet': 'internet'}, inplace=True)
    for column in tariff_stats.columns:
        tariff_stats[column + '_pct'] = tariff_stats[column] / tariff_stats[column].sum() * 100
    return tariff_stats

count_entries([messages, calls, internet])
```

Если в данных есть выбросы первого типа, их удалит функция `make_boxplot`. Она убирает из колонки все значения, которые отстоят от первого и третьего перцентилей больше, чем на полтора межквартильных интервала.

```python
def make_boxplot(df, label):
    num_rows_before = df.shape[0]
    q1 = df[label].describe()['25%']
    q3 = df[label].describe()['75%']
    upper_whisker = q3 + 1.5*(q3 - q1)
    lower_whisker = q1 - 1.5*(q3 - q1)
    if lower_whisker < 0:
        lower_whisker = 0
    if lower_whisker < df[label].min():
        lower_whisker = df[label].min()
    print('Нормальные значения: от', lower_whisker, 'до', upper_whisker)
    df = df.loc[((df[label] >= lower_whisker) & (df[label] <= upper_whisker)) | (df[label].isnull())]
    num_rows_after = df.shape[0]
    print('В датафрейме осталось', num_rows_after, 'строк')
    print('Количество строк в датафрейме сократилось на', num_rows_before - num_rows_after)
    return df  
```

От выбросов второго типа избавимся с помощью функции `sigma`, которая обрезает датафрейм по правилу трех сигм: удаляются пользователи, которые слишком часто или слишком редко звонят/шлют смс/выходят в интернет.

```python
def sigma(df):
    global users
    df_before = df.shape[0]
    users_before = users.shape[0]
    grouped = df.groupby('user_id').count()
    ceiling = grouped['id'].mean() + 3*np.std(grouped['id'])
    floor = grouped['id'].mean() - 3*np.std(grouped['id'])
    if floor < 0:
        floor = 0
    if floor < grouped['id'].min():
        floor = grouped['id'].min()
    too_little = list(grouped.loc[grouped['id'] < floor].index)
    too_much = list(grouped.loc[grouped['id'] > ceiling].index)
    df = df.query('user_id != @too_little and user_id != @too_much')
    users = users.query('user_id != @too_little and user_id != @too_much')
    df_after = df.shape[0]
    users_after = users.shape[0]
    print('Количество строк в датафрейме сократилось на', df_before - df_after)
    print('Количество пользователей в датафрейме сократилось на', users_before - users_after)
    return df

# def box_and_whiskers(df):
#     before = df.shape[0]
#     grouped = df.groupby('user_id').count()
#     q1 = grouped['id'].describe()['25%']
#     q3 = grouped['id'].describe()['75%']
#     upper_whisker = q3 + 1.5*(q3 - q1)
#     lower_whisker = q1 - 1.5*(q3 - q1)
#     if lower_whisker < 0:
#         lower_whisker = 0
#     if lower_whisker < grouped['id'].min():
#         lower_whisker = grouped['id'].min()
#     too_little = list(grouped.loc[grouped['id'] < lower_whisker].index)
#     too_much = list(grouped.loc[grouped['id'] > upper_whisker].index)
#     print('Слишком общительные:', too_much)
#     print('Слишком молчаливые:', too_little)
#     df = df.query('user_id != @too_little and user_id != @too_much')
#     after = df.shape[0]
#     print('Количество строк в датафрейме сократилось на', before - after)
#     return df
```

Поскольку цены на трафик указаны в гигабайтах, в таблицах с данными о тарифах и количестве использованного трафика создадим отдельные столбцы, где данные будут посчитаны в этих единицах.

```python
internet['gb_used'] = internet['mb_used'] / 1024
tariffs['gb_included'] = tariffs['mb_included'] / 1024
```

В `users` добавим информацию из `tariffs`.

```python
users = users.merge(tariffs, left_on='tariff', right_on='tariff_name', how='left').drop('tariff', axis=1)
```

Как мы знаем, многие пользователи подключились к "Мегалайну" или отключились от него в течение года. Для расчета среднемесячной прибыли нам нужно будет знать, сколько месяцев каждый из них был клиентом компании на момент выгрузки.

```python
users['months_on_board'] = (users['churn_date'] - users['reg_date']) / np.timedelta64(1, 'M')
```

При помощи `pivot_table` посчитаем, сколько каждый из пользователей потратил ресурсов (минут, сообщений, трафика) в каждом месяце.

```python
def create_pivot(df, col, agg):
    return df.pivot_table(col, ['user_id', 'month'], aggfunc=agg)

calls_pivot = create_pivot(calls, 'duration', 'sum')
internet_pivot = create_pivot(internet, 'gb_used', 'sum')
messages_pivot = create_pivot(messages, 'id', 'count')
```

Теперь обогатим каждую из сводных таблиц данными о помесячной прибыли при помощи функции `find_revenue`, которой в качестве аргументов передаются название сводной таблицы и два столбца из `tariffs`: с информацией о включенном объеме услуг и плате за каждую единицу сверх лимита.

```python
def find_revenue(df_pivot, prepaid, fee):
    df_pivot = df_pivot.merge(users.set_index('user_id'), how='left', left_index=True, right_index=True)    
    df_pivot['extra'] = df_pivot.iloc[:, 0] - df_pivot[prepaid]
    df_pivot['extra_fee'] = df_pivot['extra'].where(df_pivot['extra'] >= 0, 0) * df_pivot[fee]

    
    def replace_negatives(value): # функция для замены отрицательных значений нулями
        try:
            if value < 0:
                return 0
            else:
                return value
        except:
            return value
    
    
    df_pivot = df_pivot.applymap(lambda x: replace_negatives(x)) 
    return df_pivot


calls_pivot = find_revenue(calls_pivot, 'minutes_included', 'rub_per_minute')
internet_pivot = find_revenue(internet_pivot, 'gb_included', 'rub_per_gb')
messages_pivot = find_revenue(messages_pivot, 'messages_included', 'rub_per_message')
```

Склеим все три сводные таблицы и посчитаем, сколько прибыли в среднем за месяц принес компании каждый из клиентов, результат запишем в `average_monhtly_fee`.

```python
revenue_table = calls_pivot.loc[:, ['rub_monthly_fee', 'extra_fee']].merge(
    internet_pivot.loc[:, 'extra_fee'].to_frame(), how='left', left_index=True, right_index=True, suffixes=('_clls', '_intt')).merge(
    messages_pivot.loc[:, 'extra_fee'].to_frame(), how='left', left_index=True, right_index=True)

revenue_table['total_fee'] = revenue_table.sum(axis=1)
revenue_table = revenue_table.groupby('user_id').sum()
revenue_table['average_monthly_fee'] = revenue_table['total_fee'] / users.set_index('user_id')['months_on_board']
revenue_table
```

Вычислим и запишем в `users` среднемесячный расход трафика, минут и сообщений на одного пользователя каждого из двух тарифов.

```python
users = users.set_index('user_id').join(
    [calls_pivot.groupby('user_id')['duration'].sum().to_frame(),
     internet_pivot.groupby('user_id')['gb_used'].sum().to_frame(), 
     messages_pivot.groupby('user_id')['id'].sum().to_frame()]
)

users['minutes_used_monthly_mean'] = users['duration'] / users['months_on_board']
users['gb_used_monthly_mean'] = users['gb_used'] / users['months_on_board']
users['messages_sent_monthly_mean'] = users['id'] / users['months_on_board']
```

Потом прицепляем информацию о тарифах и включённых минутах и всё. Остальные расчёты будут простыми поколоночными операциями. Код станет проще и читабельнее. Проверим получившиеся сводные таблицы на предмет выбросов. В прошлый раз мы фильтровали отдельные сессии, а теперь имеем дело с агрегированными данными по месяцам -- здесь могли появиться новые аномалии, если у каких-то пользователей максимальная частота пользования услугой совпала с максимальным объемом используемых ресурсов за 1 сессию.

```python
def volume_sigma(df, label):
    num_rows_before = df.shape[0]
    ceiling = df[label].mean() + 3*np.std(df[label])
    floor = df[label].mean() - 3*np.std(df[label])
    if floor < 0:
        floor = 0
    if floor < df[label].min():
        floor = df[label].min()
    print('Нормальные значения: от', floor, 'до', ceiling)
    df = df.loc[((df[label] >= floor) & (df[label] <= ceiling)) | (df[label].isnull())]
    num_rows_after = df.shape[0]
    print('В датафрейме осталось', num_rows_after, 'строк')
    print('Количество строк в датафрейме сократилось на', num_rows_before - num_rows_after)
    return df  
```

```python
users = volume_sigma(users, 'minutes_used_monthly_mean')
```

```python
users = volume_sigma(users, 'gb_used_monthly_mean')
```

```python
users = volume_sigma(users, 'messages_sent_monthly_mean')
```

Построим графики средних среднемесячных значений для каждого тарифа и посчитаем дисперсию, среднее значение и стандарнтное отклонение. Отдельно можно построить плотностные диаграммы (они закоментированы в коде ниже).

```python
def count_var_nad_std():
    for column in users.loc[:, 'minutes_used_monthly_mean':'messages_sent_monthly_mean'].columns:
        var_smart = np.var(users.query('tariff_name == "smart"')[column])
        std_smart = np.std(users.query('tariff_name == "smart"')[column])
        mean_smart = users.query('tariff_name == "smart"')[column].mean()
        var_ultra = np.var(users.query('tariff_name == "ultra"')[column])
        std_ultra = np.std(users.query('tariff_name == "ultra"')[column])
        mean_ultra = users.query('tariff_name == "ultra"')[column].mean()
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 10))
        users.query('tariff_name == "smart"')[column].plot(kind='hist', bins=30, title=column + '_smart', ax=ax1)
        users.query('tariff_name == "ultra"')[column].plot(kind='hist', bins=30, title=column + '_ultra', ax=ax2)
#         users.query('tariff_name == "smart"')[column].plot.density(title=column + '_smart', ax=ax1)
#         users.query('tariff_name == "ultra"')[column].plot.density(title=column + '_ultra', ax=ax2)

        plt.show()
        print('Для тарифа "Smart" дисперсия составила {:.2f}, стандартное отклонение -- {:.2f}, среднее -- {:.2f}.'.format(
            var_smart, std_smart, mean_smart))    
        print('Для тарифа "Ultra" дисперсия составила {:.2f}, стандартное отклонение -- {:.2f}, среднее -- {:.2f}.'.format(
            var_ultra, std_ultra, mean_ultra))    
count_var_nad_std()
```

Сразу заметно, что данные по тарифу "Smart" более однородные, чем по "Ultra", что подтверждается б**о**льшими значениями дисперсии и стандартного отклонения в первом случае. Возможно, это связано с тем, что "Ultra" хуже представлена в нашей выборке. Если напрячь воображение, то и там можно увидеть очертания нормального распределения. По средним значениям можно судить о том, сколько ресурсов каждого типа требуется пользователям. Перейдем к проверке гипотез.

Нулевая гипотеза 1: "Средние выручки пользователей тарифов "Ultra" и "Smart" равны".
Альтернативная гипотеза 1: "Средние выручки пользователей тарифов "Ultra" и "Smart" различаются". Посмотрим, можно ли отвергнуть нулевую гипотезу 1 с традиционным уровнем статистической значимости в 5%. Поскольку есть предположение, что обе выборки взяты из одной и той же генеральной совокупности, параметр `equal_var` трогать не будем.

```python
sample_1 = users.query('tariff_name == "ultra"')['average_monthly_fee'].dropna()
sample_2 = users.query('tariff_name == "smart"')['average_monthly_fee'].dropna()

results = st.ttest_ind(sample_1, sample_2)

print('p-значение: ', results.pvalue)
```

Можно. Вероятность, что различия выручек -- результат случайности, не просто не превышает пороговое значение в 5%, а стремится к нулю. При данном размере выборок той разницы в среднемесячном чеке между пользователями двух тарифов, которую мы получили в начале этого раздела, с лихвой хватает для того, чтобы утверждать, что эта разница не случайна, а связана именно с использованием двух разных тарифов.

Нулевая гипотеза 2: "Средние выручки пользователей из Москвы и из других регионов равны".
Альтернативная гипотеза 2: "Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов".
Уровень достоверности снова примем равным 5%.

```python
sample_1 = users.query('city == "Москва"')['average_monthly_fee'].dropna()
sample_2 = users.query('city != "Москва"')['average_monthly_fee'].dropna()

results = st.ttest_ind(sample_1, sample_2)

print('p-значение: ', results.pvalue)
```

Вторую нулевую гипотезу отвергнуть нельзя, различия в выручке между Москвой и не-Москвой статистически не значимы.
